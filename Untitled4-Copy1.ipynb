```python
import pandas as pd
import numpy as np

projectdataframe = pd.read_csv('legislators-current.csv')
projectdataframe
```

```python
projectdataframe.info()
```

```python
#Drop middlename, suffix
dataframe = projectdataframe.drop(columns=projectdataframe.columns[2:5])
dataframe
#Drop all columns after party affiliation
newdataframe = dataframe.drop(columns=dataframe.columns[10:])
newdataframe
#Drop class
verynewdataframe = newdataframe.drop(columns=newdataframe.columns[7:9])
verynewdataframe
```

```python
#Add age column by birthday functio that creates age
from datetime import datetime, date


# This function converts given date to age
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month,
                                      today.day) < (born.month,
                                                    born.day))

verynewdataframe['Age'] = verynewdataframe['birthday'].apply(age)
verynewdataframe
```

```python
verynewdataframe["Age"].describe()
```

```python
Agebyhouse=verynewdataframe[["type", "Age"]].groupby("type").mean()
Agebyhouse
```

```python
Agebygender=verynewdataframe[["gender", "Age"]].groupby("gender").mean()
Agebygender
```

```python
Agebyparty=verynewdataframe[["party", "Age"]].groupby("party").mean()
Agebyparty
```
